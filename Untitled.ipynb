```python
# nbi:hide_in
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import nbinteract as nbi
import ipywidgets as widgets
from datetime import date
import datetime
from IPython.display import display, HTML
df=pd.read_csv('https://raw.githubusercontent.com/jmsevillam/Calculadoras/master/ESPECIETF.csv',';',parse_dates=True)
df['Emision']=pd.to_datetime(df['Emision'], format='%d/%m/%Y')
df['Vencimiento']=pd.to_datetime(df['Vencimiento'], format='%d/%m/%Y')
def trunc(num,n=3):
        return np.floor(10.**n * num) / 10.**n
    
def prev_pay(today,date_Ven):
    '''Calculates the date of the previous payment'''
    date_Ant=date_Ven.replace(year=today.year)
    if ((date_Ant-today)/np.timedelta64(1,'D')>0):
        date_Ant=date_Ant.replace(year=today.year-1)
    return date_Ant

def next_pay(today,date_Ven):
    '''Calculates the date of the next payment'''
    date_Next=date_Ven.replace(year=today.year+1)
    if today.month < date_Ven.month:
        date_Next=date_Next.replace(year=today.year)
    elif today.month==date_Ven.month:
        if today.day<=date_Ven.month:
            date_Next=date_Next.replace(year=today.year)
    return date_Next

def Duration(today,date_Ven,cupon,tasa):
    '''Calculates the Duration of the bond'''
    Dmac=0.
    div=0.
    pago=1.

    date_Next=next_pay(today,date_Ven)
    date_Ant=prev_pay(today,date_Ven)
    pay_year=date_Next.year+1
    pagos=date_Ven.year-today.year
    nu=((today-date_Ant).days)/365.
    for pago in range(1,1+pagos):
        if(pay_year==date_Ven.year): cupon+=1.
        peso=round((cupon*100.)*((1./(1.+tasa)**(nu))*(1./(1.+tasa)**(pago-1))),1)
        Dmac+=(pago)*peso
        div+=peso
        pay_year+=1
    return trunc(Dmac/div)

def PSucio(date_Ven,today,cupon,tasa):
    '''Calculates the price "Sucio"'''
    date_Next=next_pay(today,date_Ven)
    date_Cupon=date_Ven
    year=date_Next.year
    date_prev=prev_pay(today,date_Ven)
    suma=0.
    bic=0.
    n=0
    while year<=date_Ven.year:
        if (year%4==0 and year%100!=0 and today.month>2) or (year%4==0 and year%100!=0 and today.month==2 and today.day>28): bic+=1
        if year==date_Ven.year: cupon+=1
        
        date_Cupon=date_Cupon.replace(year=year)
        year+=1
        #print(date_prev,today)
        if n==0:
            Dias=((date_Next-today)/np.timedelta64(1,'D'))-bic
            n=(Dias/365.)
        vp1=trunc(cupon*100./((1.+tasa)**(n)),3)
        suma+=vp1        
        n+=1.
        
    return trunc(suma)

def Dias_Vto(today,date_Ven):
    date_Next=next_pay(today,date_Ven)
    year=date_Next.year
    bic=0.        
    while year<=date_Ven.year:
        if (year%4==0 and today.month>2) or (year%4==0 and today.month==2 and today.day>28): bic+=1
        year+=1
    Dias=((date_Ven-today)/np.timedelta64(1,'D'))-bic
    return int(Dias)

def interests(cupon,today,date_Ven,column):
    fut=0.
    if column=='D': fut=1.
    '''Calculates the interests of the bond'''
    date_Ant=prev_pay(today,date_Ven)
    inte=(((cupon)*100.)/365.)*(int((today-date_Ant)/np.timedelta64(1,'D'))+1-fut)
    return round(inte,6) 



def Tase(Vencimiento,Hoy,Cupon,Limpio):
    Intereses=interests(Cupon,Hoy,Vencimiento,'D')
    Sucio=Limpio+Intereses
    print(Intereses,Limpio,Sucio)
    an=0.
    bn=1.
    eps=0.000001
    sucios=[]
    while (np.abs(an-bn)>eps):
        pn=(an+bn)/2.
        Test1=PSucio(Vencimiento,Hoy,Cupon,an)-Sucio
        Test2=PSucio(Vencimiento,Hoy,Cupon,pn)-Sucio
        #print('aa',Test1+Sucio,Test2+Sucio,Sucio,Intereses)
        if((Test1)*(Test2)>0):
            an=pn
        else:
            bn=pn
    return pn


'''Buttons and widgets'''

a=widgets.FloatText(
    value=7.5,
    description='Tasa Real:',
    disabled=False
)
b=widgets.FloatText(
    value=1000000000,
    description='Nominal:',
    disabled=False
)
c=widgets.FloatText(
    value=0,
    description='Futuros:',
    disabled=False
)
a2=widgets.FloatText(
    value=7.5,
    description='Tasa Real:',
    disabled=False
)
b2=widgets.FloatText(
    value=1000000000,
    description='Nominal:',
    disabled=False
)
c2=widgets.IntText(
    value=0,
    description='Futuros:',
    disabled=False
)

a3=widgets.FloatText(
    value=100.,
    description='Precio Limpio:',
    disabled=False
)
b3=widgets.FloatText(
    value=1000000000,
    description='Nominal:',
    disabled=False
)
c4=widgets.IntText(
    value=0,
    description='Futuros:',
    disabled=False
)
a4=widgets.FloatText(
    value=100.,
    description='Precio Limpio:',
    disabled=False
)
b4=widgets.FloatText(
    value=1000000000,
    description='Nominal:',
    disabled=False
)
c4=widgets.IntText(
    value=0,
    description='Futuros:',
    disabled=False
)
```
